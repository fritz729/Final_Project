```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#read csv and create a dataframe
df = pd.read_csv('Resources/Data_science_&stem_salaries.csv')
df.head()
```

```python
salaryDF = df[['company', 'totalyearlycompensation', 'yearsofexperience', 'yearsatcompany', 'basesalary']]
salaryDF.head()
```

```python
y = df.totalyearlycompensation, df.basesalary
x = df.drop(columns="totalyearlycompensation") 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    random_state=1,
                                                    stratify=y)
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='ibfgs',
                                 max_iter=200,
                                 random_state=1)
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual":y_test}).rest_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
